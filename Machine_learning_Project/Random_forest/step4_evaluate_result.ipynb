```python
import numpy
import pandas as pd
import joblib
import json
import seaborn

import sklearn

%matplotlib inline
```

### Loading testing data & Model

```python
data_test = pd.read_csv('../data/salary.test.processed.csv').set_index('id')
data_test
```

```python
model =joblib.load('./model/model.joblib')
model
```

### Calculate and add `prediction` column

```python
model.predict(X = data_test.drop(['label'],axis='columns'))
```

```python
data_test['prediction']=model.predict(
X = data_test.drop(columns = ['label']),
)
data_test
```

```python
data_test.to_csv('./results/predictions.test.csv')
```

---

### Calculate confusion metrix

```python
sklearn.metrics.confusion_matrix(
    y_true=data_test['label'],
    y_pred=data_test['prediction'],
)
```

```python
report_scores = sklearn.metrics.classification_report(
    y_true=data_test['label'],
    y_pred=data_test['prediction'],
    digits = 6,
    output_dict = True
)
df_score = pd.DataFrame(report_scores).transpose()
df_score
```

```python
with open('./results/scores.test.json','w')as f:
    json.dump(
        obj=report_scores,
        fp=f,
        indent = 4
    )
```
